```python
import glob
import pandas as pd

df_list = [pd.read_csv(f) for f in glob.glob('result_v4/*.csv')]
final_df = pd.concat(df_list,ignore_index=True)
final_df['return:mdd'] = final_df['overall']/(-1*final_df['mdd'])
```

```python
final_df['overall_R'] = final_df['overall'].rank(method='average')
final_df['winrate_R'] = final_df['winrate'].rank(method='average')
final_df['mdd_R'] = final_df['mdd'].rank(method='average')
final_df['expectency_R'] = final_df['expectancy'].rank(method='average')
final_df['return:mdd_R'] = final_df['return:mdd'].rank(method='average')
final_df['total_R'] = final_df.iloc[:, -5:].sum(axis=1)
```

```python
monday_0 = final_df[final_df['best_day/w'] == 0].sort_values('total_R')

tuesday_1 = final_df[final_df['best_day/w'] == 1].sort_values('total_R')

wednesday_2 = final_df[final_df['best_day/w'] == 2].sort_values('total_R')

thursday_3 = final_df[final_df['best_day/w'] == 3].sort_values('total_R')

friday_4 = final_df[final_df['best_day/w'] == 4].sort_values('total_R')
```
