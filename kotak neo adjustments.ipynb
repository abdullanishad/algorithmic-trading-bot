```python
# creating nfodf 
cashUrl = client.scrip_master(exchange_segment = "NSE")
nfoUrl = client.scrip_master(exchange_segment = "NFO")
nfodf = pd.read_csv(nfoUrl)
nfodf.columns = [c.strip() for c in nfodf.columns.values.tolist()]
nfodf['lExpiryDate'] = pd.to_datetime(nfodf['lExpiryDate'],unit='s').apply(lambda x: x.date() + relativedelta(years=10) ) 
```

```python
# filtering nfodf based on underlying and expiry date
nfodf['lExpiryDate'] = pd.to_datetime(nfodf['lExpiryDate'])
desired_date = pd.to_datetime(expiry)
nfodf = nfodf[(nfodf.pSymbolName == index) & (nfodf.lExpiryDate == desired_date)]
nfodf
```

```python
def adjust(adjustment,opt_type):
    # creating custom position_df 
    position_df = pd.DataFrame(client.positions()['data'])
    position_df = position_df[['buyAmt','prod','flBuyQty','flSellQty','tok','sellAmt','stkPrc','trdSym','optTp',]]
    position_df['qty'] = position_df.flBuyQty.astype(int) - position_df.flSellQty.astype(int)
    row = position_df[(position_df.optTp == opt_type) & (position_df.qty < 0)]

    if not row.empty:
        # existing strike data
        existing_strike_ltp = round(float(client.quotes(instrument_tokens=[{"instrument_token": row.tok.iloc[0], "exchange_segment": "nse_fo"}],
                                                        quote_type="ltp", isIndex=False)['message'][0]['ltp']),2)
        qty = str(abs(row.qty.iloc[0]))
        buy_orderid = client.place_order(exchange_segment="NFO", product="NRML", price=str(existing_strike_ltp), order_type="L",
                           quantity=qty, validity="DAY", trading_symbol=row.trdSym.iloc[0],
                           transaction_type="B", amo="NO", disclosed_quantity="0",
                           market_protection="0",pf="N",trigger_price="0", tag=None)
        if buy_orderid['stat'] == 'Ok':
            print('check')
            order_report = pd.DataFrame(client.order_report()['data'])
            sleep(1)
            while order_report[(order_report.nOrdNo == buy_orderid['nOrdNo'])].ordSt.iloc[0] == 'open':
                print('modifying open buy order.....')
                existing_strike_ltp += 0.05
                existing_strike_ltp = round(existing_strike_ltp,2)
                buy_orderid = client.modify_order(order_id = buy_orderid['nOrdNo'], 
                                                  price = str(existing_strike_ltp), quantity = qty,
                                                  trigger_price = "0", validity = "DAY",
                                                  order_type = "L", amo = "")
                order_report = pd.DataFrame(client.order_report()['data'])
                sleep(1)
            print('buy order placed')

            # new strike data
            if adjustment == 'risk_on' and opt_type == 'CE':
                new_strike = (float(row['stkPrc'].iloc[0]) - 50) *100
            elif adjustment == 'risk_off' and opt_type == 'CE':
                new_strike = (float(row['stkPrc'].iloc[0]) + 50) *100
            elif adjustment == 'risk_on' and opt_type == 'PE':
                new_strike = (float(row['stkPrc'].iloc[0]) + 50) *100
            elif adjustment == 'risk_off' and opt_type == 'PE':
                new_strike = (float(row['stkPrc'].iloc[0]) - 50) *100
                
            new_strike_symbol = nfodf[(nfodf['dStrikePrice;'] == new_strike) & (nfodf.pOptionType == opt_type)]['pTrdSymbol'].iloc[0]
            new_strike_ltp = round(float(client.quotes(instrument_tokens=[{"instrument_token": row.tok.iloc[0], "exchange_segment": "nse_fo"}], 
                                                quote_type="ltp", isIndex=False)['message'][0]['ltp']),2)
            sell_orderid = client.place_order(exchange_segment="NFO", product="NRML", price=str(new_strike_ltp), order_type="L",
                               quantity=qty, validity="DAY", trading_symbol=new_strike_symbol,
                               transaction_type="S", amo="NO", disclosed_quantity="0",
                               market_protection="0",pf="N",trigger_price="0", tag=None)
            if sell_orderid['stat'] == 'Ok':
                order_report = pd.DataFrame(client.order_report()['data'])
                while order_report[(order_report.nOrdNo == sell_orderid['nOrdNo'])].ordSt.iloc[0] == 'open':
                    print('modifying open sell order.....')
                    new_strike_ltp -= 0.05
                    new_strike_ltp = round(new_strike_ltp,2)
                    sell_orderid = client.modify_order(order_id = sell_orderid['nOrdNo'], 
                                                      price = str(new_strike_ltp), quantity = qty,
                                                      trigger_price = "0", validity = "DAY",
                                                      order_type = "L", amo = "")
                    order_report = pd.DataFrame(client.order_report()['data'])
                print('sell order placed')
```
